```python
# Import dependencies.
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

## 1. Line Charts

### MATLAB Method

```python
# Create the plot using plt.show()
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="d", color="green", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid and a legend
plt.grid()
plt.legend()
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

### Object-Oriented Approach

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

## 3. Scatter Plots and Bubble Charts

### MATLAB Method

```python
# Creating a scatter plot using "plt.plot"
plt.plot(x_axis, y_axis, 'o')
```

```python
#  Creating a scatter plot using "plt.scatter"
plt.scatter(x_axis, y_axis)
```

```python
# ANNOTATE CHART
# Create the plot, with x and y flipped:
plt.scatter(y_axis, x_axis, color="red", label='Chicago')
# Create labels for the x and y axes.
plt.ylabel("Date")
plt.xlabel("Fare($)")
# Inverting the order of the y axis, so that it goes from Jan-Dec:
plt.gca().invert_yaxis()
# Add the legend.
plt.legend()
# Create a title.
plt.title("PyBer Fare by Month")
```

```python
# Creating a scatter plot with different size bubbles, according to the fare prices (y_axis):
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
# Iterate through prices in y_axis and multiply them by 3 to make bubbles bigger:
y_axis_larger = []
for data in y_axis:
  y_axis_larger.append(data*3)
```

```python
# Creating a scatter plot using list with fares multiplied by 3:
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
# Using list comprehension instead of the for loop:
plt.scatter(x_axis, y_axis, s=[i*3 for i in y_axis])
```

### Object-Oriented Approach

```python
# Create a Scatter Plot Using the Object-Oriented Interface
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, s=y_axis)
```

```python
# ANNOTATE CHART
# Create the plot, with x and y flipped:
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, 
           c="skyblue", 
           alpha=0.8,
           #mec='black',
           #mew=2,
           s=[i*5 for i in y_axis], 
           label="Boston")
# Add a title to the graph
ax.set_title('PyBer Fare by Month')
# Create labels for the x and y axes.
ax.set_xlabel('Fare($)')
ax.set_ylabel('Date')
# Invert the y-axis so that January is at the top
ax.invert_yaxis()
# Add legend
ax.legend()
```

## 4. Pie Charts

### MATLAB Method

```python
# Create a Pie Chart Using the MATLAB Method
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
# MATLAB approach:
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", 
          "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
# Increase the size of the chart, so it's more visible:
plt.subplots(figsize=(8, 8))
# Create explode_values tuple to make one of the pie wedges pop
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
# Plot pie chart
plt.pie(y_axis, explode=explode_values, colors=colors, labels=x_axis, autopct='%.1f%%')
plt.show()
```

### Object-Oriented Approach

```python
# Object-oriented approach:
# Assign 12 colors, one for each month.
colors = ["slateblue", "thistle", "lightblue", "cadetblue", "olivedrab", "yellowgreen", 
          "gold", "orange", "palevioletred", "indianred", "lightcoral", "mistyrose"]
fig, ax = plt.subplots(figsize=(8, 8))
# Create explode_values tuple to make one of the pie wedges pop
explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
# Plot chart
ax.pie(y_axis, labels=x_axis, colors=colors, autopct='%.1f%%', 
       explode=explode_values, shadow=True, startangle=265)
# Invert the y-axis so that January is at the top
ax.invert_yaxis()
plt.show()
```

## 5. Error Bars

```python
import statistics
```

```python
# Get the standard deviation of the values in the y-axis.
stdev = statistics.stdev(y_axis)
stdev
```

### MATLAB Method

```python
# MATLAB approach:
plt.errorbar(x_axis, y_axis, yerr=stdev, capsize=3)
```

```python
# Add Error Bars to a Bar Chart (MATLAB approach):
plt.bar(x_axis, y_axis, yerr=stdev, capsize=3)
```

### Object-Oriented Approach

```python
# Object-oriented approach:
fig, ax = plt.subplots()
ax.errorbar(x_axis, y_axis, yerr=stdev, capsize=3)
plt.show()
```

```python
# Add Error Bars to a Bar Chart (Object-oriented approach):
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis, yerr=stdev, capsize=3)
plt.show()
```

## 6. Major Ticks

### MATLAB Method

```python
# MATLAB approach
# Set major ticks on x axis to every $5 increment
import numpy as np
plt.barh(x_axis, y_axis)
plt.xticks(np.arange(0, 51, step=5.0))
plt.gca().invert_yaxis()
```

### Object-Oriented approach

```python
# Object-oriented approach:
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis)
ax.set_xticks(np.arange(0, 51, step=5.0))
plt.show()
```

## 7. Minor Ticks

### Object-Oriented Approach

```python
# Add Minor Ticks (Object-oriented):
from matplotlib.ticker import MultipleLocator
# Increase the size of the plot figure.
fig, ax = plt.subplots(figsize=(8, 8))
ax.barh(x_axis, y_axis)
ax.set_xticks(np.arange(0, 51, step=5.0))

# Create minor ticks at an increment of 1.
ax.xaxis.set_minor_locator(MultipleLocator(1))
plt.show()
```
